```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
alpha,beta,r,s = 0.006,0.00025,1.0015,0.9994
x0 = 2
y0 = 3
rounds = 1000000
X = [x0]
Y = [y0]
```

```python
for i in range(rounds):
    x = r * X[i] - alpha * X[i] * Y[i]
    y = s * Y[i] + beta * X[i] * Y[i]
    X.append(x)
    Y.append(y)
```

```python

plt.figure(figsize = (15,15))
plt.plot(X,label = 'Prey')
plt.plot(Y, label = 'Predator')
plt.legend()
```
